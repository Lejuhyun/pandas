## 결측 데이터 필터링

```python
cond = df_copy['age'].isnull() #결측 데이터 필터링
df_copy.loc[cond, 'age'] = 30
df_copy.tail()
```

## fillna()

```python
df_copy = df.copy()
```

```python
df_copy.tail()
```

```python
df_copy['age'] = df_copy['age'].fillna(100) # 재할당해야댐
```

```python
df_copy.tail()
```

```python
df_copy['deck'].value_counts()
```

```python
df_copy['deck']= df_copy['deck'].cat.add_categories('Z')
df_copy['deck'].fillna('Z')
```

```python
age_mean = df_copy['age'].mean()
df_copy['age'] = df_copy['age'].fillna(age_mean)
```

```python
df_copy.info()
```

```python
df_copy = df.copy()
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].median())
df_copy.info()
```

```python
df_copy.tail()
```

## practice

- 남자승객의 결측치는 남자승객의 평균

- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df_copy = df.copy()
```

```python
df_copy.tail(10)
```

```python
cond1 = df_copy['sex'] == 'male'
cond1_mean= df_copy[cond1]['age'].mean()
```

```python
df_copy['age'] = df_copy.loc[cond1,'age'].fillna(cond1_mean)
```

```python
df_copy
```

```python
cond2 = df_copy['sex'] == 'female'
```

```python
cond2_mean = df_copy[cond2]['age'].mean()
```

```python
df_copy['age'] = df_copy.loc[cond2]['age'].fillna(cond2_mean)
```

```python
df_copy['age'].mean()
```

```python
# 선생님

df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
```

```python
male_mean = df_copy.loc[cond1, 'age'].mean()
print(male_mean)
```

```python
cond2 = df_copy['sex'] == 'female'
female_mean = df_copy.loc[cond2,'age'].mean()
print(female_mean)
```

```python
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2,'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
