## 자전거 데이터 활용

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head(2)
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'],format ='%b-%d-%Y')
# %b : Jan, Feb ...
df_copy.info()
```

```python
cond = df_copy['대여일자'].dt.dayofweek == 3
df_copy.loc[cond]
```

```python
df_copy.head(3)
```

```python
pd.to_numeric(df_copy['운동량'])
```

```python
df_copy.loc[2344]
```

```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors= 'coerce')  # errors = ['ignore, 'raise','coerce']
```

```python
df_copy.info()
```

## cut()

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins)
```

```python
pd.cut(df_copy['운동량'], bins = 10) #가장 작은 데이터부터 가장 큰 데이터를 10개의 구간으로 나눈다
pd.cut(df_copy['운동량'], bins = 10).value_counts()
```

```python
#qcut: 비율로 나눈다
pd.qcut(df_copy['운동량'],q=10)
pd.qcut(df_copy['운동량'],q=10).value_counts()
```

```python
qcut_bins = [0, 0.20, 0.8, 1]
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'],qcut_bins, labels= ['적음', '보통', '많음'])
df_copy.head(2)
```
