# Where()
- 해당하는 조건이 False인 애들을 바꿔줌

```python
cond = df['fare'] < 30
df['fare'].where(cond, 1).tail()  # => 'fare' >= 30인 애들을 일괄적으로 1로 바꿔줌
```

## Practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

- 남성 흡연자 조건 필터링
- 팁을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending= False).head(10)
```

- 식사인원이 5명 이상인 저녁 테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sun','Sat'])
tips.loc[(cond1 & cond2) & (cond3 & cond4),['total_bill', 'tip']]
```
