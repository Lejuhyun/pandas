# 1. 인덱싱

- iloc => 정수 인덱스로 인덱싱, 마지막 인덱스 번호는 포함하지 않음
- loc => 문자열로 인덱싱, 마지막 인덱스 번호 포함
- 열 인덱싱 => 칼럼이름으로
- 행 인덱싱 => 인덱스 이름으로

## <iloc (숫자로 인덱싱, 끝번호 포함 X)>
1. 행만 가져오기
- df.iloc[ 인덱스번호 ]
- df.iloc[ [ 인덱스번호 여러개 ] ]
- df.iloc[ 인덱스번호 슬라이싱 ]
  
2. 열만 가져오기
- df.iloc[ : , 칼럼번호]
- df.iloc[ : , [ 칼럼번호 여러개 ] ]
- df.iloc[ : , 칼럼번호 슬라이싱 ]
  
3. 행과 열 가져오기 
- df.iloc[ 행, 열 ]
- df.iloc[ 0, 0 ]
- df.iloc[ [1, 3], [0, 2, 3] ]
- df.iloc[ 2:5, 1:3 ]

## <loc (라벨, 숫자로 인덱싱, 끝번호 포함)>

1. 행만 가져오기
- df.loc[ 인덱스번호 ]
- df.loc[ [ 인덱스번호 여러개 ] ]
- df.loc[ 인덱스번호 슬라이싱 ]


2. 열만 가져오기
- df.loc[ '칼럼이름' ]
- df.loc[ ['칼럼1', '칼럼2'] ]
- df.loc[ '칼럼1' : '칼럼2' ]

3. 행과 열 가져오기
- df.loc[ '라이츄', 1 ]
- df.loc[[ '리자드', '꼬부기'], [1,2,3]]
- df.loc[ '라이츄': '꼬부기', 1:3]

**여러개 행을 가져올때만 대괄호이용, 슬라이싱할때는 대괄호 필요 없음**


```python
df.loc[2:5, ['pclass','who']] # 2~5번째 행에서 'pclass'와 'who' 칼럼을 가져오기
```

```python
df.iloc[2:6, [1,9]] 
# 1. iloc는 슬라이싱 마지막 값을 포함하지 않는다
# 2. iloc는 str로 접근할 수 없다, 숫자로만 접근 가능
```

```python
df.loc[[3,17],'pclass':'who']  # 3번과 17번 행에서 'pclass'부터 'who' 행을 가져오기
```

# 8. 결측치
1. 결측데이터 확인
   - df.isnull()
   - df.notnull()
2. 결측데이터 채우기 **재할당해야된다**
   - df.fillna()
3. 결측치 제거하기
    - df.dropna(how='any')
    - how : {'any' : 존재하면 제거 / 'all' : 모두 결측치면 제거}

```python
df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
# 결측치 확인하기
df_copy.isnull().sum()
```

```python
# age의 null값을 평균으로 재할당하기
age_mean = df_copy['age'].mean()
```

```python
df_copy['age'] = df_copy['age'].fillna(age_mean)
```

```python
df_copy.info()
```

```python
df_copy.dropna()
```

# practice
- 남자승객의 결측치는 남자승객의 평균
- 여자승객의 결측치는 여자승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
df_copy = df.copy()
```

```python
# 남자 승객의 결측치는 남자승객의 평균
cond1 = df_copy['sex'] == 'male'
male_mean = df_copy.loc[cond1, 'age'].mean()
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean)
```

```python
# 여자 승객의 결측치는 여자 승객의 평균
cond2 = df_copy['sex'] == 'female'
female_mean = df_copy.loc[cond2,'age'].mean()
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
# 결측치를 모두 채우고 난 뒤, 모든 사람의 평균
df_copy['age'].mean()
```
